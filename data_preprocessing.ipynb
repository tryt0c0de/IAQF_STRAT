```python
import pandas as pd
```

```python
bbg_ms_df = pd.read_excel("bbg_data.xlsx", sheet_name=3, header=[3, 4, 5], index_col=0)
bbg_spx_df = pd.read_excel("bbg_data.xlsx", sheet_name=0, header=[3, 4, 5], index_col=0)
bbg_all_df = pd.read_excel("bbg_data.xlsx", sheet_name=1, header=[3, 4, 5], index_col=0)

bbg_df = pd.merge(bbg_ms_df, bbg_spx_df, left_index=True, right_index=True)

bbg_df
```

```python
tickers = ["AAPL", "AMZN", "GOOGL", "META", "MSFT", "NVDA", "TSLA", "SPX"]
bbg_columns = []
for ticker in tickers:
    bbg_columns.append(f"{ticker}_price")
    bbg_columns.append(f"{ticker}_mcap")
    bbg_columns.append(f"{ticker}_return")

bbg_df.columns = bbg_columns

bbg_df
```

```python
goog_df = bbg_all_df["GOOG UW Equity"].copy()
goog_df.columns = ["GOOG_price", "GOOG_mcap", "GOOG_return"]

goog_df
```

```python
bbg_df = pd.merge(bbg_df, goog_df, left_index=True, right_index=True)

bbg_df
```

```python
bbg_df.to_csv("bbg_data.csv")
```
